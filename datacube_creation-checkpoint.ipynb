# Pre-process the datasets

## Setting the CRS for era data

```python
# Definition of a function to print the crs of the data
def print_crs(dataray):
    """Prints the crs of the dataray"""
    return print(dataray.rio.crs)
```

```python
print_crs(era)
```

```python
print_crs(ndvi)
```

```python
# Definition of a function to define a crs for the dataray
def define_crs(dataray, crs=2154):
    """Defines a crs for the dataray"""
    return dataray.rio.write_crs(crs, inplace=True)
```

```python
# Era datasets is in WG84 or EPSG:4326
era = define_crs(era, 4326)
```

```python
print_crs(era)
```

## Defining the Area Of Interest (aoi) 

```python
# Definition of a function to define the area of interest
def define_area_of_interest(aoi):
    """Defines the area of interest"""
    "input: path to aoi's shapefile"
    crop_extent = gpd.read_file(aoi)
    return crop_extent

# Definition of a function to print the crs of the crop_extent
def print_crs_crop_extent(crop_extent):
    """Prints the crs of the crop_extent"""
    return print(crop_extent.crs)
```

```python
aoi = define_area_of_interest("mygeodata.zip")
print_crs_crop_extent(aoi)
```

## Cliping to the AOI

```python
#Function to clip a datasets to the area of interest
def clip_to_aoi(dataset, aoi):
    """Clips the dataray to the area of interest"""
    "input: dataray, aoi"
    return dataset.rio.clip(aoi.geometry.apply(mapping), aoi.crs)
```

### Era5 copernicus data

```python
# Clipping the era data set to the aoi
era_clip = clip_to_aoi(era,aoi)
```

### Density from worlpop

```python
#Clipping the density dataset to the aoi
density_clip = clip_to_aoi(density,aoi)
```

## Projection to the Lambert 93 (EPSG 2154)

```python
# Definition of a function to reproject a dataray from longlat to Lambert93
def reproject_to_lambert93(dataray):
    """Reprojects a dataray from longlat to Lambert93"""
    return dataray.rio.reproject("EPSG:2154")
```

```python
era_clip_proj = reproject_to_lambert93(era_clip)
```

```python
plt.rcParams["figure.figsize"] = [6, 6]
plt.rcParams["figure.autolayout"] = True

plt.subplot(2,1,1)
era_clip['tp'][0].plot()
plt.title("Before proj")

plt.subplot(2,1,2)
era_clip_proj['tp'][0].plot()
plt.title("After proj")



plt.show()
```

```python
density_clip_proj = reproject_to_lambert93(density_clip)
```

```python
plt.rcParams["figure.figsize"] = [6, 7]
plt.rcParams["figure.autolayout"] = True

plt.subplot(2,1,1)
density.plot()
plt.title("Before proj")

plt.subplot(2,1,2)
density_clip_proj.plot()
plt.title("After proj")



plt.show()
```

```python
def print_raster(raster):
    """Prints the raster's metadata and shape"""
    print(
        f"shape: {raster.rio.shape}\n"
        f"resolution: {raster.rio.resolution()}\n"
        f"bounds: {raster.rio.bounds()}\n"
        f"sum: {raster.sum().item()}\n"
        f"CRS: {raster.rio.crs}\n"
    )
```

```python
print('Before process\n-------')
print_raster(density)
print('After clip process\n-------')
print_raster(density_clip)
print('After clip+projection process\n-------')
print_raster(density_clip_proj)
```

## Regridding

Definition of a instance that will stand for the common grid to be shared.
The one from NDVI is already in 1km by 1km.


```python
# definition of a common grid to interpolate from a dataset to another
common_grid = rxr.open_rasterio("ndvi_01012010_09012022.nc").isel(time=0)
```

```python
# Definition of a function to interpolate the data to a common grid
def interpolate_to_common_grid(dataray):
    """Interpolates the dataray to a common grid"""
    return dataray.rio.reproject_match(common_grid)
```

```python
era_final = interpolate_to_common_grid(era_clip_proj)
```

```python
plt.rcParams["figure.figsize"] = [6, 6]
plt.rcParams["figure.autolayout"] = True

plt.subplot(2,1,1)
era_clip_proj['tp'][0].plot()
plt.title("After proj")


plt.subplot(2,1,2)
era_final['tp'][0].plot()
plt.title("After regridding")


plt.show()
```

```python
print('After clip+projection process\n-------')
print_raster(era_final['tp'][0])

print('w/o projection to ndvi\n-------')
print_raster(era_clip_proj['tp'][0])
```

## Resample to one day

```python
def resample_to_daily(dataray):
    """Resamples the dataray to daily values"""
    return dataray.resample(time="1D").interpolate("linear")
```

```python
ndvi_daily=resample_to_daily(ndvi)
```
