```python
import pickle
import pandas as pd
from explainerdashboard import ClassifierExplainer, ExplainerDashboard
```

```python
# Get the path of the needed files
path_data = '..\model\df_model.csv'
path_model = '..\model\model_lgbm.pkl'
```

```python
# Load the data into a datafram
df = pd.read_csv(path_data)
```

```python
# Split the test data
df_xtest = df.drop(['index','FireMask','time','FireMask_pred','FireMask_proba'], axis=1)
df_ytest = df['FireMask']
```

```python
#load the model
model_path = open(path_model , 'rb')
model = pickle.load(model_path)
```

```python
# Create and run the expaliner in local on port 9050
explainer = ClassifierExplainer(model, df_xtest, df_ytest)
ExplainerDashboard(explainer).run(port=9050)
```
